## Exploratory Analysis and Visualization




````python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
````

**TODO** - total  matches won by every team 

````python
data=match_df.iloc[:,10]
season =data.value_counts().keys().tolist()
matches=data.value_counts().tolist()
print(season,matches)
````

````python
import matplotlib.pyplot as plt
plt.gcf().set_size_inches(20,10)
plt.xlabel('Teams')
plt.ylabel('Matches won')
plt.bar(season, matches, color = 'c')
plt.xticks(season)
plt.show()
````

**TODO** - most toss wins by teams

````python
sub_data2 = match_df.iloc[:, 6]
teams = sub_data2.value_counts().keys().tolist()
values = sub_data2.value_counts().tolist()
print(teams, values)
````

````python
plt.gcf().set_size_inches(15,5)
plt.xlabel('toss won')
plt.ylabel('teams')
plt.barh(teams, values, color = 'm', )
plt.gca().invert_yaxis()
plt.xlim(0, 90)
plt.show()
````

**TODO** - trend of winning toss and winning matches

````python
sub_data3 = match_df.loc[:, ['toss_winner', 'winner']] 
is_true = sub_data3['toss_winner'] == sub_data3['winner']
sub_data3 = sub_data3[is_true]
sub_data5=sub_data3.shape[0]
sub_data4=match_df.shape[0]-sub_data3.shape[0]
````

````python
labels = ['Yes', 'No']
share = [sub_data5,sub_data4]
plt.pie(share, labels=labels,autopct='%1.1f%%', startangle=90)
plt.title('Toss determining Win')
my_circle=plt.Circle( (0,0), 0.3, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
````

**TODO** - top venues hosting matches

````python
matter=match_df.iloc[:,14]
venues =matter.value_counts().head(5).keys().tolist()
matches=matter.value_counts().head(5).tolist()
print(venues,matches)
````

````python
plt.gcf().set_size_inches(20,5)
plt.xlabel('venues')
plt.ylabel('Matches host')
plt.bar(venues, matches, color = 'c')
plt.show()
````

**TODO** - umpire experience

````python
sub_data6 = match_df.loc[:, ['umpire1', 'umpire2']]
umpire1 = sub_data6['umpire1'].tolist()
umpire2 = sub_data6['umpire2'].tolist()
umpires = umpire1 + umpire2
from collections import Counter
c = Counter(umpires)
umpires = []
count = []
for i,j in c.most_common(10):
    umpires.append(i)
    count.append(j)
print(umpires, count)
````

````python
plt.gcf().set_size_inches(15,5)
plt.xlabel('Umpires')
plt.ylabel('Matches')
plt.title('Experienced Umpires')
plt.bar(umpires, count)
plt.show()
````

````python
sub_data2 = match_df.iloc[:, 13]
players = sub_data2.value_counts().head(10).keys().tolist()
values = sub_data2.value_counts().head(10).tolist()
print(players, values)
````

#### Q2: TODO - which team has most wins percentage

````python
sub_data7 = match_df.loc[:, ['team1', 'team2', 'winner']]
team1 = sub_data7['team1'].tolist()
team2 = sub_data7['team2'].tolist()
teams = team1 + team2
c = Counter(teams)
teams = []
matches = []
for i,j in c.most_common():
    teams.append(i)
    matches.append(j)
wins = sub_data7['winner'].value_counts().tolist()
print(teams, matches, wins)
loss = [int(n1) - int(n2) for n1,n2 in zip(matches, wins)]
print(loss)
````

````python
plt.gcf().set_size_inches(20,10)
plt.xlabel('Team')
plt.ylabel('Wins | Matches')
plt.title('Matches won from the number of matches played')
def subcategorybar(X, vals, width=0.8):
    n = len(vals)
    _X = np.arange(len(X))
    plt.bar(_X, vals[0], width=width/n, color = '#F51F05')
    plt.bar(_X, vals[1], width=width/n, color = '#05F579')
    plt.xticks(_X, X)

subcategorybar(teams, [matches, wins])
label = ['Total', 'Won']
plt.legend(labels = label)
plt.show()
````
